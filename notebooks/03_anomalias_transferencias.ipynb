# 03 - Detección de Anomalías en Transferencias

Chatbot que analiza el concepto de transferencias financieras y detecta si son usuales o inusuales.

**Entrada:**
- `id_movimiento`: Identificador único de la transferencia
- `monto`: Monto en USD (positivo)
- `concepto`: Descripción de la transferencia (1-125 caracteres)

**Salida:**
- `resultado`: "Usual" o "Inusual"
- `razon_si_inusual`: Explicación si es inusual (None si es usual)
- `confianza`: Score de confianza (0-1)

## 1. Setup: Cargar módulos y API key

```python
import os
import sys
from dotenv import load_dotenv

# Agregar src al path para importar módulos
sys.path.insert(0, '../src')

load_dotenv(dotenv_path="../.env")

api_key = os.getenv("GOOGLE_API_KEY")
print("✓ API key cargada" if api_key else "✗ API key no encontrada")
```

## 2. Importar el analizador de transferencias

```python
from anomalies_transfer import TransferenceAnalyzer, TransferenceInput

analyzer = TransferenceAnalyzer(api_key=api_key)
print("✓ Analizador de transferencias listo")
```

## 3. Análisis individual de transferencias

```python
# Caso 1: Transferencia usual
transfer_usual = TransferenceInput(
    id_movimiento="TRX001",
    monto=1500.00,
    concepto="Pago de servicios mensuales"
)

result1 = analyzer.analyze(transfer_usual)
print(f"ID: {result1.id_movimiento}")
print(f"Resultado: {result1.resultado}")
print(f"Confianza: {result1.confianza:.1%}")
if result1.razon_si_inusual:
    print(f"Razón: {result1.razon_si_inusual}")
```

```python
# Caso 2: Transferencia con posibles red flags
transfer_suspicious = TransferenceInput(
    id_movimiento="TRX002",
    monto=50000.00,
    concepto="Transferencia a cuenta offshore sin documentación"
)

result2 = analyzer.analyze(transfer_suspicious)
print(f"ID: {result2.id_movimiento}")
print(f"Resultado: {result2.resultado}")
print(f"Confianza: {result2.confianza:.1%}")
if result2.razon_si_inusual:
    print(f"Razón: {result2.razon_si_inusual}")
```

## 4. Análisis en lote

```python
transferencias_batch = [
    TransferenceInput(id_movimiento="TRX003", monto=250.50, concepto="Compra de artículos de oficina"),
    TransferenceInput(id_movimiento="TRX004", monto=100000.00, concepto="Lavado de dinero operativo"),
    TransferenceInput(id_movimiento="TRX005", monto=3200.00, concepto="Alquiler mensual del departamento"),
    TransferenceInput(id_movimiento="TRX006", monto=75000.00, concepto="Pago a beneficiarios desconocidos en paraíso fiscal"),
]

resultados_batch = analyzer.analyze_batch(transferencias_batch)

print("\n" + "="*70)
print("ANÁLISIS EN LOTE")
print("="*70)

for res in resultados_batch:
    status = "✓" if res.resultado == "Usual" else "⚠️ "
    print(f"{status} {res.id_movimiento:8} | {res.resultado:8} | Confianza: {res.confianza:.1%}")
    if res.razon_si_inusual:
        print(f"           → {res.razon_si_inusual}")
    print()
```

## 5. Análisis interactivo

```python
# Define tu propia transferencia
custom_transfer = TransferenceInput(
    id_movimiento="TRX_CUSTOM",
    monto=5000.00,
    concepto="Pago a proveedor internacional de software"
)

result_custom = analyzer.analyze(custom_transfer)

print("📊 ANÁLISIS PERSONALIZADO")
print(f"ID: {result_custom.id_movimiento}")
print(f"Clasificación: {result_custom.resultado}")
print(f"Nivel de confianza: {result_custom.confianza:.1%}")
if result_custom.razon_si_inusual:
    print(f"Observaciones: {result_custom.razon_si_inusual}")
else:
    print("Observaciones: Transferencia dentro de patrones normales")
```
