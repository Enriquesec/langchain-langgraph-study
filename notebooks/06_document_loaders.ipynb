# 06 - Cargar Documentos en LangChain

LangChain proporciona **Document Loaders** para cargar documentos de diferentes fuentes:
- PDFs
- CSVs
- HTML
- Archivos de texto
- Páginas web
- Bases de datos
- Y muchas más...

Cada loader convierte el documento a objetos `Document` que LangChain puede procesar.

## PARTE 1: Estructura de Document

```python
from langchain.schema import Document

print("ESTRUCTURA DE UN DOCUMENT EN LANGCHAIN")
print("=" * 70)
print()

# Un Document tiene dos partes principales:
doc = Document(
    page_content="Este es el contenido del documento",
    metadata={
        "source": "archivo.pdf",
        "page": 1,
        "fecha": "2025-01-15"
    }
)

print(f"page_content: {doc.page_content}")
print(f"metadata: {doc.metadata}")
print()
print("Beneficios:")
print("✓ Contenido: Texto a procesar")
print("✓ Metadata: Información sobre el origen del documento")
print("✓ Trazabilidad: Saber de dónde vino cada chunk")
```

## PARTE 2: Cargar Archivos de Texto

```python
from langchain.document_loaders import TextLoader
import os

print("CARGAR ARCHIVO DE TEXTO")
print("=" * 70)
print()

# Crear archivo de prueba
archivo_txt = "documento_prueba.txt"
with open(archivo_txt, "w") as f:
    f.write("""LangChain es un framework para desarrollar aplicaciones con LLMs.
Permite conectar modelos de lenguaje con APIs externas.
Es de código abierto y soporta múltiples proveedores de LLMs.""")

print(f"1. Archivo creado: {archivo_txt}")
print()

# Cargar con TextLoader
loader = TextLoader(archivo_txt)
documents = loader.load()

print(f"2. Documento cargado:")
print(f"   Total documentos: {len(documents)}")
print()
print(f"3. Contenido:")
print(f"   {documents[0].page_content}")
print()
print(f"4. Metadata:")
print(f"   {documents[0].metadata}")

# Limpiar
os.remove(archivo_txt)
```

## PARTE 4: Cargar CSVs

```python
from langchain.document_loaders.csv_loader import CSVLoader
import csv
import os

print("CARGAR ARCHIVOS CSV")
print("=" * 70)
print()

# Crear archivo CSV de prueba
csv_file = "datos_prueba.csv"
with open(csv_file, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["nombre", "edad", "ciudad"])
    writer.writerow(["Juan", "30", "Madrid"])
    writer.writerow(["María", "28", "Barcelona"])
    writer.writerow(["Pedro", "35", "Valencia"])

print(f"1. Archivo CSV creado: {csv_file}")
print()

# Opción 1: CSVLoader (carga cada fila como documento)
print("Opción 1: CSVLoader (cada fila es un documento)")
print("-" * 70)
loader = CSVLoader(file_path=csv_file)
docs = loader.load()

print(f"Total documentos: {len(docs)}")
print(f"Primer documento:")
print(f"  Contenido: {docs[0].page_content}")
print(f"  Metadata: {docs[0].metadata}")
print()
print()

# Opción 2: Cargar con columna específica
print("Opción 2: Cargar CSV con columna como clave")
print("-" * 70)
loader = CSVLoader(
    file_path=csv_file,
    source_column="nombre"  # Usar columna 'nombre' en metadata
)
docs = loader.load()

print(f"Documento con source_column:")
print(f"  {docs[0].metadata}")
print()
print()

# Opción 3: Cargar con Pandas
print("Opción 3: Cargar CSV con Pandas (más flexible)")
print("-" * 70)
print()
print("import pandas as pd")
print("from langchain.schema import Document")
print()
print("df = pd.read_csv('datos.csv')")
print("documents = [")
print("    Document(")
print("        page_content=row.to_string(),")
print("        metadata={'source': 'datos.csv', 'row': i}")
print("    )")
print("    for i, row in df.iterrows()")
print("]")
print()

# Limpiar
os.remove(csv_file)
```

## PARTE 5: Cargar HTML

```python
from langchain.document_loaders import UnstructuredHTMLLoader
import os

print("CARGAR ARCHIVOS HTML")
print("=" * 70)
print()

# Crear archivo HTML de prueba
html_file = "pagina_prueba.html"
html_content = """<!DOCTYPE html>
<html>
<head>
    <title>Mi Página</title>
</head>
<body>
    <h1>Bienvenido a LangChain</h1>
    <p>LangChain es un framework increíble para aplicaciones con LLMs.</p>
    <p>Permite trabajar con múltiples proveedores de modelos.</p>
</body>
</html>"""

with open(html_file, "w") as f:
    f.write(html_content)

print(f"1. Archivo HTML creado: {html_file}")
print()

# Opción 1: UnstructuredHTMLLoader
print("Opción 1: UnstructuredHTMLLoader")
print("-" * 70)
print()
print("Instalación:")
print("  pip install unstructured[html]")
print()
print("Uso:")
print("  from langchain.document_loaders import UnstructuredHTMLLoader")
print("  loader = UnstructuredHTMLLoader('pagina.html')")
print("  docs = loader.load()")
print()
print()

# Opción 2: BeautifulSoup (manual)
print("Opción 2: Cargar HTML con BeautifulSoup (control total)")
print("-" * 70)
print()
print("Instalación:")
print("  pip install beautifulsoup4")
print()
print("Uso:")
print("  from bs4 import BeautifulSoup")
print("  from langchain.schema import Document")
print()
print("  with open('pagina.html', 'r') as f:")
print("      soup = BeautifulSoup(f, 'html.parser')")
print("      # Extraer texto limpio")
print("      text = soup.get_text()")
print("      doc = Document(")
print("          page_content=text,")
print("          metadata={'source': 'pagina.html'}")
print("      )")
print()
print()

# Opción 3: HTMLSectionLoader (divide por secciones)
print("Opción 3: HTMLSectionLoader (divide por HTML tags)")
print("-" * 70)
print()
print("from langchain.document_loaders import UnstructuredHTMLLoader")
print()
print("loader = UnstructuredHTMLLoader('pagina.html')")
print("# Divide automáticamente por <h1>, <h2>, <section>, etc")
print("docs = loader.load()")
print()

# Limpiar
os.remove(html_file)
```

## PARTE 7: Cargar Múltiples Archivos

```python
from langchain.document_loaders import DirectoryLoader
import os

print("CARGAR MÚLTIPLES ARCHIVOS DE UN DIRECTORIO")
print("=" * 70)
print()

# Crear archivos de prueba
os.makedirs("docs_prueba", exist_ok=True)
for i in range(3):
    with open(f"docs_prueba/documento{i}.txt", "w") as f:
        f.write(f"Este es el documento {i} sobre LangChain")

print("1. Directorio creado con 3 archivos .txt")
print()

# Cargar todos los .txt
print("Opción 1: Cargar todos los TXT")
print("-" * 70)
print()
loader = DirectoryLoader(
    "docs_prueba",
    glob="**/*.txt",  # Patrón: todos los .txt
    use_multithreading=True  # Cargar en paralelo
)
docs = loader.load()

print(f"Total documentos cargados: {len(docs)}")
for i, doc in enumerate(docs):
    print(f"  Doc {i}: {doc.metadata['source']}")
print()
print()

print("Opción 2: Cargar con diferentes loaders")
print("-" * 70)
print()
print("from langchain.document_loaders import TextLoader")
print()
print("loader = DirectoryLoader(")
print("    'docs/',")
print("    glob='**/*.txt',")
print("    loader_cls=TextLoader")
print(")")
print("docs = loader.load()")
print()
print()

print("Opción 3: Cargar múltiples tipos de archivos")
print("-" * 70)
print()
print("from langchain.document_loaders import PDFLoader, TextLoader, CSVLoader")
print()
print("documents = []")
print()
print("# Cargar PDFs")
print("pdf_loader = DirectoryLoader('docs/', glob='**/*.pdf', loader_cls=PDFLoader)")
print("documents.extend(pdf_loader.load())")
print()
print("# Cargar TXTs")
print("txt_loader = DirectoryLoader('docs/', glob='**/*.txt', loader_cls=TextLoader)")
print("documents.extend(txt_loader.load())")
print()
print("# Cargar CSVs")
print("csv_loader = DirectoryLoader('docs/', glob='**/*.csv', loader_cls=CSVLoader)")
print("documents.extend(csv_loader.load())")
print()

# Limpiar
import shutil
shutil.rmtree("docs_prueba")
```

## PARTE 8: Mejores Prácticas

```python
print("MEJORES PRÁCTICAS PARA CARGAR DOCUMENTOS")
print("=" * 70)
print()

print("1. SIEMPRE LIMPIA EL CONTENIDO")
print("-" * 70)
print()
print("from langchain.text_splitter import CharacterTextSplitter")
print()
print("# Cargar")
print("docs = loader.load()")
print()
print("# Dividir en chunks")
splitter = CharacterTextSplitter(")
print("chunks = splitter.split_documents(docs)")
print()
print("# Cada chunk es un Document independiente")
print()
print()

print("2. MANTÉN METADATA RELEVANTE")
print("-" * 70)
print()
print("# Buena metadata")
print("metadata = {")
print("    'source': 'documento.pdf',")
print("    'page': 1,")
print("    'date': '2025-01-15',")
print("    'author': 'John Doe'")
print("}")
print()
print("# Permite filtrar y rastrear documentos después")
print()
print()

print("3. MANEJA ERRORES EN CARGA MASIVA")
print("-" * 70)
print()
print("import glob")
print()
print("documents = []")
print("for file in glob.glob('docs/**/*.pdf', recursive=True):")
print("    try:")
print("        loader = PyPDFLoader(file)")
print("        docs = loader.load()")
print("        documents.extend(docs)")
print("    except Exception as e:")
print("        print(f'Error cargando {file}: {e}')")
print("        continue")
print()
print()

print("4. NORMALIZA CONTENIDO")
print("-" * 70)
print()
print("import re")
print()
print("def limpiar_documento(doc):")
print("    # Remover espacios múltiples")
print("    text = re.sub(r'\\s+', ' ', doc.page_content)")
print("    # Remover caracteres especiales")
print("    text = re.sub(r'[^\\w\\s]', '', text)")
print("    doc.page_content = text.strip()")
print("    return doc")
print()
print("cleaned_docs = [limpiar_documento(d) for d in documents]")
print()
print()

print("5. CACHEA DOCUMENTOS CARGADOS")
print("-" * 70)
print()
print("import pickle")
print()
print("# Guardar")
print("with open('documentos_cache.pkl', 'wb') as f:")
print("    pickle.dump(documents, f)")
print()
print("# Cargar desde cache")
print("with open('documentos_cache.pkl', 'rb') as f:")
print("    documents = pickle.load(f)")
print()
print("# Evita recargar archivos cada vez")
```

## PARTE 10: Ejemplo Completo RAG con Documentos Reales

```python
print("EJEMPLO COMPLETO: Cargar Documentos + RAG")
print("=" * 70)
print()

print("Paso 1: Cargar Documentos")
print("-" * 70)
print("from langchain.document_loaders import PDFLoader, TextLoader")
print("from langchain.document_loaders import DirectoryLoader")
print()
print("# Cargar PDFs")
print("pdf_loader = DirectoryLoader('documentos/', glob='**/*.pdf', ")
print("                            loader_cls=PDFLoader)")
print("pdf_docs = pdf_loader.load()")
print()
print("# Cargar TXTs")
print("txt_loader = DirectoryLoader('documentos/', glob='**/*.txt',")
print("                            loader_cls=TextLoader)")
print("txt_docs = txt_loader.load()")
print()
print("# Combinar")
print("all_docs = pdf_docs + txt_docs")
print()
print()

print("Paso 2: Dividir en Chunks")
print("-" * 70)
print("from langchain.text_splitter import RecursiveCharacterTextSplitter")
print()
print("splitter = RecursiveCharacterTextSplitter(")
print("    chunk_size=1000,")
print("    chunk_overlap=100")
print(")")
print()
print("chunks = splitter.split_documents(all_docs)")
print()
print()

print("Paso 3: Crear Vector Store")
print("-" * 70)
print("from langchain_google_genai import GoogleGenerativeAIEmbeddings")
print("from langchain_community.vectorstores import FAISS")
print()
print("embeddings = GoogleGenerativeAIEmbeddings(model='models/embedding-001')")
print("vector_store = FAISS.from_documents(chunks, embeddings)")
print()
print()

print("Paso 4: Crear RAG Chain")
print("-" * 70)
print("from langchain.chains import RetrievalQA")
print("from langchain_google_genai import ChatGoogleGenerativeAI")
print()
print("llm = ChatGoogleGenerativeAI(model='gemini-2.5-flash')")
print()
print("qa_chain = RetrievalQA.from_chain_type(")
print("    llm=llm,")
print("    chain_type='stuff',")
print("    retriever=vector_store.as_retriever()")
print(")")
print()
print()

print("Paso 5: Hacer Consultas")
print("-" * 70)
print("pregunta = 'Cual es la política de devoluciones?'")
print("respuesta = qa_chain.run(pregunta)")
print("print(respuesta)")
print()
print()

print("BENEFICIOS:")
print("-" * 70)
print("✓ Carga de múltiples formatos (PDF, TXT, CSV, HTML)")
print("✓ Metadata preservada para rastrear documentos")
print("✓ Búsqueda semántica rápida")
print("✓ Respuestas precisas basadas en documentos reales")
print("✓ Fácil de mantener y actualizar")
```
